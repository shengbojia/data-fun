```python
import numpy as np
import hashlib


# Problem is everytime you run this you are going to get a different training/test split
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

def test_set_check(identifier, test_ratio, hash):
    # Hash the id, use the last byte of the hash to determine whether or not to put in test set
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    # Apply ids as id_ in the anonmy func
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    
    # in_test_set is boolean array
    return data.loc[~in_test_set], data.loc[in_test_set]

# Our dataset does not have identifier column, so just use row index
housing_with_id = housing.reset_index() # Adds 'index' column
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'index') # 20% will be test set
```

```python
# The above was simple random sampling, but since out dataset is not extremely large, sampling bias might occur

# Lets try Stratified sampling, with stratas formed from median income (as this column is very importatnt for predicting prices)

# Most median income between 2 - 5 (thousand)
housing["income_cat"] = np.ceil(housing["median_income"] / 1.5) # shrink it by 1.5 to limit number of strata, 
# and ceil to make it dsicreet
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace=True) # group > 5 into 5
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
    
housing["income_cat"].value_counts() /len(housing) # See the distribution
    
```

```python
# Now drop the 'income_cat' column to revert back to original data
for set in (strat_train_set, strat_test_set):
    set.drop(["income_cat"], axis=1, inplace=True)
    
# Put away the test set and make a protective copy of the training set to mess around with
housing = strat_train_set.copy()
plt.rcParams['figure.figsize'] = [15, 10]
# Now lets visualize the data
housing.plot(kind="scatter", x="longitude", y="latitude")
```

```python
# Problem  here is that algorithms usually asume consecutive numbers are similar, but 0 and 1 are very different in this case
from sklearn.preprocessing import OneHotEncoder

# Lets use 'one hot encoding' instead. Vectors where one entry is 1 and others 0. eg [1 0 0 0 0] for <1H, [0 1 0 0 0] for INLAND
encoder = OneHotEncoder()
housing_cat_1hot = encoder.fit_transform(housing_cat) # Rshape as this fit_transform takes 2D arrays
housing_cat_1hot # Will be in a sparese matrix t save space (only keep track of location of the 1)
```

```python
# Make custom transformer and use it with FunctionTransformer

from sklearn.preprocessing import FunctionTransformer

rooms_ix, bedrooms_ix, population_ix, household_ix = [
    list(housing.columns).index(col)
    for col in ("total_rooms", "total_bedrooms", "population", "households")]

def add_extra_features(X, add_bedrooms_per_room=True):
    rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
    population_per_household = X[:, population_ix] / X[:, household_ix]
    if add_bedrooms_per_room:
        bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
        return np.c_[X, rooms_per_household, population_per_household,
                     bedrooms_per_room]
    else:
        return np.c_[X, rooms_per_household, population_per_household]
    
    
# Use it like so:
attr_adder = FunctionTransformer(add_extra_features, validate=False, # validate is False since data contains non float values
                                 kw_args={"add_bedrooms_per_room": False})
housing_extra_attribs = attr_adder.fit_transform(housing.values)
        

housing_extra_attribs
```

```python
# Lets create a PipeLine to facilitate the data flow

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler # for feature scaling (standarization)

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')), # Pipe line constructor takes in a list of name/estimator pairs
    ('attribs_adder', FunctionTransformer(add_extra_features, validate=False)),
    ('std_scaler', StandardScaler()) # All but last parameter must be a transformer ie have fit_transform
]) 

housing_num_tr = num_pipeline.fit_transform(housing_numerical) # fir_transform calls all the fit_transform in steps
housing_num_tr
```

```python
# Use ColumnTransformer to apply transforms/pipelines to different columns

from sklearn.compose import ColumnTransformer

num_attribs = list(housing_numerical) # List of attribs in housing_numerical
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs) # Transform ocean_proximity into one-hot vectors
])

print(housing.shape)
housing_prepared = full_pipeline.fit_transform(housing)
housing_prepared
```

```python
housing_prepared.shape
```

```python
# Time to train
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
print(housing_prepared.shape)
print(housing_labels.size)
lin_reg.fit(housing_prepared, housing_labels)
```

```python
# Lets test it out

some_data = housing.iloc[:5]
some_data.info()
some_labels = housing_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)

print(some_data_prepared.shape) # Making sure the pipeline did the correct transforms

print("Predictions:", lin_reg.predict(some_data_prepared))
print("Labels:", list(some_labels))
```

```python
# works, but now lets measure the RMSE (root mean square error)

from sklearn.metrics import mean_squared_error

housing_predictions = lin_reg.predict(housing_prepared) # Use predict on the full training set now
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse)

from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(housing_labels, housing_predictions)
print(lin_mae)
```

```python
# That's a pretty bad RMSE (remember it's the std dev, and most houses are between $120 000 to $265 000, so $68 628 error is bad)

# Underfitting right now, so lets try a more complex model, a Decision tree
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(housing_prepared, housing_labels)
```

```python
housing_predictions = tree_reg.predict(housing_prepared)
tree_mse = mean_squared_error(housing_labels, housing_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
# 0 seems unlikely, might be overfitting the training data
# Let's try some cross validation

from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, housing_prepared, housing_labels, 
                         scoring="neg_mean_squared_error", cv=10) # Split into 10 distinct 'folds'

tree_rmse_scores = np.sqrt(-scores) # SKlearn cross requires a utitlity func (maximize value) so neg-mse is used instead

# Let's display the results
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Std devivaion:", scores.std())
    
display_scores(tree_rmse_scores)
```

```python
# That actually appears worse than the linear regression
# To make sure, lets do cross-validation on the linear model as well

lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels, 
                            scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)

display_scores(lin_rmse_scores)
```

```python
# It appears that the decision tree is overfitting so badly that it underperforms the linear model
# Lets try a Random forest regressor

from sklearn.ensemble import RandomForestRegressor # Takes the mean of many decision trees on random subsets of the data

forest_reg = RandomForestRegressor(n_estimators=10, random_state=42)
forest_reg.fit(housing_prepared, housing_labels)
```

```python
housing_predictions = forest_reg.predict(housing_prepared)
forest_mse = mean_squared_error(housing_labels, housing_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
# Looks promising, but could still be overfitting the data so lets try cross-validation

forest_scores = cross_val_score(forest_reg, housing_prepared, housing_labels, 
                               scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```

```python
# 52583 is a lot better than both the linear and the single decision tree

# But observe how forest_rmse is way less than the cross-val score, so there's  still overfitting happening
# At this point we could either fine tune the random  forest, or even explore new models

# say you wanted to fiddle with different combinations of hyperparameters for the model at hand
# Lets use Grid Search to do so
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]}, # first dict has 3 x 4 = 12 combos of hyperparameters
    {'n_estimators': [3, 10], 'max_features': [2, 3, 4], 'bootstrap':[False]} # has 2 x 3 x 1 = 6 combos 
] # For a total of 12 + 6 = 18 combos

forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=5, # Will train each model(combo) 5 times, so 18 x 5 = 90 times total
                           scoring="neg_mean_squared_error") # Measures against this performance

grid_search.fit(housing_prepared, housing_labels)

```
